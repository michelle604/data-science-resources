```python

# Set the random seed to keep the example consistent.
np.random.seed(100)

# Sample data.
plt.subplot(1, 2, 1)
latitude = np.random.normal(47.4999061, 6.73096127, 213)
plt.title('latitude distribution of countries in Europe')
plt.ylabel('latitude')

# Generate and display the box plot.
plt.boxplot(latitude)
plt.show()

# Set the random seed to keep the example consistent.
np.random.seed(100)

# Sample data.
plt.subplot(1, 2, 2)
longtitude = np.random.normal(15.37431925, 13.0680136, 213)
plt.title('longtitude distribution of countries in Europe')
plt.xlabel('longtitude')

# Generate and display the box plot.
plt.boxplot(longtitude,vert=False)
plt.show()
```
